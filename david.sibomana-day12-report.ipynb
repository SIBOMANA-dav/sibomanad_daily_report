### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.


````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, 10)
y=x
plt.figure(figsize=(7, 6))
plt.plot(x, y, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````


### Exercise 2:
Plot `y = x^2` using Matplotlib for `x` values between -10 and 10.

````python
x = np.linspace(-10, 10)
y=x**2
plt.figure(figsize=(10, 8))
plt.plot(x, y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

### Exercise 3:
Create a scatter plot with 50 random points where x and y values are between 0 and 10.

````python
x=np.random.randint(10, size=50)
y=np.random.randint(10, size=50)
plt.scatter(x, y,)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

### Exercise 4:
Plot a sine wave using Matplotlib for `x` values between 0 and 2π.

````python
x=np.linspace(0, 2 *np.pi, 100)
s=np.sin(x)
plt.figure(figsize=(8,6))
plt.plot(x, s, 'b')
plt.xlabel('x')
plt.ylabel('s')
plt.show()
````

### Exercise 6:
Create a histogram of 1000 random numbers generated from a normal distribution.

````python
x=np.random.normal(loc=0, scale=1, size=1000)
plt.hist(x)
plt.xlabel('x')
plt.ylabel('frequency')
plt.show()
````

### Exercise 7:
Plot two line plots on the same graph: `y1 = x^2` and `y2 = x^3` for `x` ranging from 0 to 10.

````python
x=np.linspace(0, 10)
y_1=x**2
y_2=x**3
plt.figure(figsize=(8, 6))
plt.plot(x, y_1,  'b')
plt.plot(x, y_2)
plt.xlabel('x')
plt.ylabel('y')
#plt.ylabel('z')
plt.show()
````

### Exercise 8:
Add titles, axis labels, and a legend to the plot in Exercise 7.

````python
x=np.linspace(0, 10)
y_1=x**2
y_2=x**3
plt.figure(figsize=(8, 6))
plt.plot(x, y_1,  'b', label=r"$y=x**2$")
plt.plot(x, y_2, 'r', label=r"$y=x**3$")
plt.xlabel('time')
plt.ylabel('velocity')
#plt.ylabel('z')
plt.title(r'plot of $y=x^2$')
plt.legend(loc='upper right', frameon=False)
plt.show()
````

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

````python
x = np.linspace(0, 10)
y_1 = x**2
y_2 = x**3
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1) 
plt.plot(x, y_1, 'b')
plt.xlabel('x')
plt.ylabel('y_1')
plt.title('y_1 = x^2')
````

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

````python
x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2
y3 = x**3
plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'r-', label='y1 = x')
plt.plot(x, y2, 'b--', label='y2 = x^2')
plt.plot(x, y3, 'g:', label='y3 = x^3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y1 = x, y2 = x^2, and y3 = x^3')
````

### Exercise 11:
Create a pie chart with 5 segments labeled A, B, C, D, and E, with values `[25, 35, 20, 15, 5]`.

````python
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [25, 35, 20, 15, 5]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
````

### Exercise 12:
Create a horizontal bar chart with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`.


````python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
plt.figure(figsize=(8, 6))
plt.barh(categories, values, color='skyblue')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.tight_layout()
plt.show()
````
